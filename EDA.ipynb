```python
from datasets import load_metric, load_from_disk, Dataset, DatasetDict

datasets = load_from_disk("../data/train_dataset")
datasets
```

```python
datasets['train'].features.keys()
```

```python
import pandas as pd

index_list = list(datasets['train'].features.keys())[1:]
index_list
```

```python
pd_config = {
  'answer_start': [],
  'answer_text': []  
}
for value in index_list:
  if value == "answers":
    continue
  pd_config[value] = datasets['train'][value]
answers = list(datasets['train']['answers'])
for answer in answers:
  pd_config['answer_start'].append(answer['answer_start'][0])
  pd_config['answer_text'].append(answer['text'][0])
  if len(answer['text']) != 1:
    print(answer['text'])
train_dataset = pd.DataFrame(pd_config)
train_dataset
```

```python
train_dataset.to_csv('train.csv',index=False)
```

# 데이터 분석
- context가 중복인 문서보다 title이 중복인 문서가 더많다.
- 잘 살펴보니 title이 중복이여도 document_id가 다를 경우가 있다.
- 그럼 document_id가 연속적인 것을 합쳐서 새로운 context를 제작하고 데이터를 추가적으로 늘릴 수 있지 않을까?
- 그럼 반대로 answer_start를 이용해서 중요한 부분만 자르도록 할 수 있지 않을까?
- 맞춤법을 한번 거치는 건 어떨까?
- context 길이를 전부 가져와서 적당한 max_length를 만들자

```python
from kospellpy import spell_init
spell_checker = spell_init()
for context in train_dataset['context']:
  total_context = spell_checker(context)
  if total_context != context:
    print(total_context)
    print(context)
    break
```
