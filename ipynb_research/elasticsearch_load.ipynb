```python
datasets = load_from_disk("/opt/ml/data/new_train_dataset")
```

```python
datasets['train'][0]
```

```python
k = 10
score = 0
correct_set = []
correct_examples = []
wrong_set = []
for data in datasets['train']:
  query = data['question']
  scores, retrieved_examples = wiki_datasets.get_nearest_examples("context", query, k=k)
  if data['context'] in retrieved_examples['context']:
    correct_set.append(data)
    correct_examples.append(retrieved_examples)
    score += 1
  else: 
    wrong_set.append(data)
```

```python
correct_set[3], correct_examples[3]
```

```python
score/len(datasets['train']) * 100
```
